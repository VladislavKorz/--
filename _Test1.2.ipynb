```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import numpy as np

date_birth = datetime.datetime(1998, 6, 11)

df = pd.read_csv(r"brent-month_csv.csv", sep=',', parse_dates=['Date'])

df = df[df['Date'] > date_birth]
print("Среднее цены на нефть", df['Price'].sum() / df['Price'].count())

print("Минимальная цена на нефть", min(df['Price']))
print("Максимальная цена на нефть", max(df['Price']))
print("Cреднеквадратическое отклонение цены на нефть", np.std(df['Price']))

df_plt = df[df['Date'] > datetime.datetime.today()-datetime.timedelta(days=365)]
plt.plot(df_plt['Date'], df_plt['Price'])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import numpy as np
import seaborn as sbn

date_birth = datetime.datetime(1998, 6, 11)

df = pd.read_csv(r"crypto_data_updated_13_november.csv", sep=',')
for item in list(df):
    if not 'Close' in item:
        df.drop([item], axis=1, inplace=True)

sbn.pairplot(df)
plt.show()


plt.figure(figsize=(12,10), dpi= 80)
sbn.heatmap(df.corr(), xticklabels=df.corr().columns, 
            yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)

plt.title('Карта корреляций', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
